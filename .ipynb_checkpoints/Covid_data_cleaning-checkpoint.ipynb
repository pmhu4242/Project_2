```python
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np
```

```python
Adata = pd.read_csv("USA.csv")
Adata.head()
```

```python
#checking columns names
Adata.columns
```

```python
#Re-organized columns 
filled = Adata[["state", "totalTestResults", "totalTestsAntibody", "recovered", "positive","positiveTestsAntibody"]]
filled.head()
```

```python
#Finding average of hit, run and strikeouts for each team 

Covid_positive= filled.groupby("state").positive.mean()
Positive_Antibody= filled.groupby("state").positiveTestsAntibody.mean()
Total_Covid_Test= filled.groupby("state").totalTestResults.mean()
Total_Antibody_Test= filled.groupby("state").totalTestsAntibody.mean()
Covid_recovered= filled.groupby("state").recovered.mean()

B_Analysis = pd.DataFrame({"Covid_positive":(Covid_positive),
                        "Positive_Antibody":(Positive_Antibody),
                        "Total_Covid_Test":(Total_Covid_Test),
                        "Total_Antibody_Test":(Total_Antibody_Test),
                        "Covid_recovered":(Covid_recovered)
                        })
B_Analysis= B_Analysis.sort_values("state",ascending=True)
B_Analysis["Covid_positive"] = B_Analysis["Covid_positive"].astype(float).map("{:.2f}".format)
B_Analysis["Positive_Antibody"] = B_Analysis["Positive_Antibody"].astype(float).map("{:.2f}".format)
B_Analysis["Total_Covid_Test"] = B_Analysis["Total_Covid_Test"].astype(float).map("{:.2f}".format)
B_Analysis["Total_Antibody_Test"] = B_Analysis["Total_Antibody_Test"].astype(float).map("{:.2f}".format)
B_Analysis["Covid_recovered"] = B_Analysis["Covid_recovered"].astype(float).map("{:.2f}".format)
B_Analysis.head(5)
```

```python
#Fill nan values with 0
Nan_filled= B_Analysis.replace('NaN','0')
Nan_filled.head(5)
```
