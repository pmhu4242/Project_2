```python
import pandas as pd
import requests
from bs4 import BeautifulSoup

url = 'https://www.macrotrends.net/stocks/charts/WYND/wyndham-destinations,-inc/revenue'

r = requests.get(url)
soup = BeautifulSoup(r.content, 'html.parser')

print(soup)
```

```python
table = soup.find_all('table')[0] 
table
```

```python
df = pd.read_html(str(table))[0]
df
```

```python
df.to_csv (r"~/Desktop/Project_2/CSV's/wyndham_data\wyndham_ann_rev_dataframe.csv", index = False, header=True)
```
