```python
import pandas as pd
import requests
from bs4 import BeautifulSoup

url_1 = 'https://www.tsa.gov/coronavirus/passenger-throughput?page=0'
url_2 = 'https://www.tsa.gov/coronavirus/passenger-throughput?page=1'
# dfs = pd.read_html(url_1)

r = requests.get(url_1)
soup = BeautifulSoup(r.content, 'html.parser')

print(soup)
```

```python
table = soup.find_all('table')[0] 
table
```

```python
df = pd.read_html(str(table))[0]
df.head()
```

```python
r2 = requests.get(url_2)
soup2 = BeautifulSoup(r2.content, 'html.parser')

print(soup2)
```

```python
table2 = soup2.find_all('table')[0] 
table2
```

```python
df2 = pd.read_html(str(table2))[0]
df2.head()
```

```python
new = pd.concat([df, df2])
new
```

```python
new.to_csv (r"~/Desktop/Project_2/CSV's\tsa_dataframe.csv", index = False, header=True)
```
